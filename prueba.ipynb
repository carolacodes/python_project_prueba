```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Crear un DataFrame de ejemplo
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo"],
    "Ventas": [1200, 1500, 1800, 1700, 2200]
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print("DataFrame de prueba:")
display(df)

# 2. Calcular estadísticas simples
print("\nPromedio de ventas:", df["Ventas"].mean())
print("Máxima venta:", df["Ventas"].max())
print("Mínima venta:", df["Ventas"].min())

# 3. Graficar
plt.figure(figsize=(8, 4))
plt.plot(df["Mes"], df["Ventas"], marker="o", linestyle="solid", color="purple")
plt.title("Ventas por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas")
plt.grid(True)
plt.show()
```
